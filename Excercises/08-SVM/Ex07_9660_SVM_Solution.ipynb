### Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

### Set global variable

```python
global RANDOM_STATE
RANDOM_STATE=42
```

### Load data

```python
# Use column 'id' as the index column
df = pd.read_csv('https://raw.githubusercontent.com/vjavaly/Baruch-CIS-9660/main/data/healthcare_stroke_2.csv',
                 index_col='id')
```

### Prepare data

### Use the SimpleImputer to replace missing values

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df['bmi'] = imp_mean.fit_transform(df[['bmi']])
imp_most_frequent = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df['smoking_status'] = imp_most_frequent.fit_transform(df[['smoking_status']]).ravel()
```

### Encode categorical variables 'gender', 'ever_married' and 'smoking_status'

```python
df = pd.get_dummies(df, columns=['gender', 'ever_married', 'smoking_status'], dtype=int)
```

```python
# Display first few rows of updated dataframe
df.head()
```

### Separate independent and dependent variables
Dependent variable: 'stroke'

```python
X = df.drop('stroke', axis=1)     # Independent variables
y = df['stroke']                  # Dependent variable
```

### Standardize independent variables

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Split data into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y,
                                                    test_size=0.3,
                                                    random_state=RANDOM_STATE)
```

### Train SVC model (with default hyperparameters)

```python
model = SVC(random_state=RANDOM_STATE)
model.fit(X_train, y_train)
```

### Evaluate SVC model

```python
model_preds = model.predict(X_test)
model_accuracy = accuracy_score(y_test, model_preds)
print(f"Model score: {round((model_accuracy * 100), 5)}%")
```

### Train SVC model (with at least 2 different hyperparameters)

```python
model = SVC(C=0.5, kernel='linear', class_weight={0: 0.95, 1: 0.05},
            gamma='auto', random_state=RANDOM_STATE)
model.fit(X_train, y_train)
```

### Evaluate updated SVC model

```python
model_preds = model.predict(X_test)
model_accuracy = accuracy_score(y_test, model_preds)
print(f"Model score: {round((model_accuracy * 100), 5)}%")
```
