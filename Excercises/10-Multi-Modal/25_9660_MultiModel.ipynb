### Import libraries

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### Load data

```python
bc_data = load_breast_cancer()
df = pd.DataFrame(data=bc_data['data'], columns=bc_data['feature_names'])
df['target'] = bc_data['target']
```

### Examine data

```python
pd.set_option('display.max_columns', None)
```

### Prepare data

```python
# Check for missing values
df.isnull().sum()
```

```python
# Display class distribution
df['target'].value_counts()
```

### Separate independent and dependent variables
* Independent variables: All remaining variables except 'target'
* Dependent variable: 'target'

```python
X = df.drop("target", axis = 1)
y = df["target"]
```

### Define global variables

```python
global TEST_SIZE, RANDOM_STATE
TEST_SIZE = .3
RANDOM_STATE = 42
```

#### Standardize the independent variables

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Split data into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y,
                                                    test_size=TEST_SIZE,
                                                    random_state=RANDOM_STATE)
```

### Train Logistic Regression model

```python
LogReg_model = LogisticRegression(random_state=RANDOM_STATE)
LogReg_model.fit(X_train, y_train)
```

### Evaluate Logistic Regression model

```python
LogReg_predictions = LogReg_model.predict(X_test)
print(f'Logistic regression accuracy = {round(accuracy_score(y_test, LogReg_predictions)*100,2)}%')
```

### Train Random Forest Classifier model

```python
RF_model = RandomForestClassifier(random_state=RANDOM_STATE)
RF_model.fit(X_train, y_train)
```

### Evaluate Random Forest Classifier model

```python
RF_predictions = RF_model.predict(X_test)
print(f'Random forest accuracy = {round(accuracy_score(y_test, RF_predictions)*100,2)}%')
```
